# How I Met Your Mother

**How I Met Your Mother** (*HIMYM*) is an American sitcom that aired on CBS from September 19, 2005, to March 31, 2014. The series follows the main character, **Ted Mosby**, and his group of friends in New York City's Manhattan. As a framing device, **Ted**, in the year 2030, recounts to his son and daughter the events that led him to meet their mother.

### Main Characters

- **Ted Mosby** (*Josh Radnor*)
- **Marshall Eriksen** (*Jason Segel*)
- **Lily Aldrin** (*Alyson Hannigan*)
- **Robin Scherbatsky** (*Cobie Smulders*)
- **Barney Stinson** (*Neil Patrick Harris*)

The show was created by **Carter Bays** and **Craig Thomas**.

```python
# Example viewership data
data = {
    'Season': [1, 2, 3, 4, 5, 6, 7, 8, 9],
    'Viewership (millions)': [9.47, 8.37, 7.65, 7.81, 9.42, 8.77, 8.50, 7.80, 9.05]
}

df = pd.DataFrame(data)

# Plot the viewership over time
plt.figure(figsize=(10, 6))
plt.plot(df['Season'], df['Viewership (millions)'], marker='o', color = "red")
plt.title('How I Met Your Mother Viewership Over Time')
plt.xlabel('Season')
plt.ylabel('Viewership (millions)')
plt.xticks(df['Season'])
plt.grid(True)
plt.show()
```

### Viewership Analysis

The viewership of **How I Met Your Mother** experienced several changes over its nine-season run:

- The highest viewership was in **Season 1** with an average of **9.47 million** viewers.
- The viewership decreased to **7.65 million** by **Season 3**.
- Between **Seasons 3 and 5**, the viewership increased by **1.77 million**, reaching **9.42 million** in **Season 5**.
- There was a slight decline in viewership after **Season 5**, dropping to **7.80 million** in **Season 8**.
- The final season (**Season 9**) saw a rebound in viewership to **9.05 million**.

These fluctuations can be attributed to various factors, including storyline developments, cast changes, and competition from other TV shows.
